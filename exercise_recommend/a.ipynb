```python
import pandas as pd
import numpy as np

# Generate dummy data
np.random.seed(42)

# Create a DataFrame with 100 rows
data = pd.DataFrame({
    'age': np.random.randint(18, 60, 100),
    'height': np.random.randint(150, 190, 100),
    'sex': np.random.choice(['female', 'male'], 100),
    'days_to_goal': np.random.randint(30, 180, 100),
    'activity_level': np.random.randint(1, 6, 100),
    'experience': np.random.randint(1, 6, 100)
})

# Assign a dummy workout_plan
# This is just for illustration; you should replace it with your actual workout plan data
data['workout_plan'] = np.random.choice(['Plan A', 'Plan B', 'Plan C'], 100)

# Save the DataFrame to a CSV file
data.to_csv('workout_plan.csv', index=False)
```
